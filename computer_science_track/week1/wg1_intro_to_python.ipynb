#### Different types for different functionality

```python
# Even though Python does a lot of 'smart stuff' for you, you'll still need to be explicit.
# For example:

print('2 + 4')
print(2 + 4)
```

If you ran the cell above you'll see that there's a difference in output for Python. That is because it considers anything within parentheses as a so-called 'string'. Basically, Python considers it to be a word or sentence, so it'll keep it as is and not perform any operations on it, even if you put these operations in the sentence (like in the example above).

Now if you do not include the quotation marks, Python will not see the typed information as a sentence. Therefore, the '+' will be executed and the result of the operation is 6.

#### Equal signs and negations

```python
# Three very often used signs shall be demonstrated in this cell.

number = 9 # we saw this one before, this is an assignment. It makes the thing on the left equal to the thing on the right.
# In this case, the variable becomes equal to 9
print(number)

print(number == 9) # This is a check, to see whether the two things are the same, without replacing/altering a value
# It results in a True/False output based on whether the two things are each other's equal or not.

print(number != 9) # This is the opposite of the previous statement, it checks whether two things are NOT the same

# The '!' sign is amazing, it can negate almost anything and is often used.
# Look, it also works for strings!
print("Hello world" != "hello world")
```

#### For, while, if :)

```python
# Sometimes you want to perform something a specific amount of times:
print('___________for loop example:____________')# just to ensure clarity in the output beneath this cell

for x in range(6):
    print(x)  # Notice the spaces at the beginning of the line.
    # That is because Python uses indentation to indicate a block of code!!!

# BEWARE: most programming languages will count from 0, as you can also see in the output of the for loop.

print('__________second for loop example____________')
# NOTE: You are NEVER restricted with these, you can use as many for loops in for loops as you need :)
animals = ["chicken", "cow", "cat"]
for word in animals:
    for letter in word:
        print(letter, word)
```

```python
# Sometimes you only want to perform an action if a certain condition is met:
print('_____________If statement example:_____________') # just to ensure clarity in the output beneath this cell

if number < 20: # notice how I use the variable of the previous cell? Cool huh
    print('the number is currently smaller than 20')
if number > 20:
    print('something must have gone wrong, the number is suppose to be smaller than 20')
# You can play around with this by redefining the value of 'number', try it and you'll see :)

# Or, you might want to do something if a condition is met, and something else if the condition is not met:
print('\n_____________If else statement example:_____________') # just to ensure clarity in the output beneath this cell
if number % 2 == 0:
    print('the number is even')
else:
    print('the number is odd')

# Sometimes we want to combine the two, and perform an action while a certain condition is met :)
# IMPORTANT, if you do not use this properly you'll get an infinite loop and your laptop will NOT like it. So be careful
print('\n___________While loop example:_______________') # just to ensure clarity in the output beneath this cell

while number > 0:
    print(number)
    number = number - 1 # This step is necessary because otherwise you'll never break out of the loop!\
# please ONLY use this if you understand why I used the 'number = number - 1'. Otherwise stay clear of this
# Question to test if you understand: How come running this cell for a second time will give VERY different results?
```

#### Importing libraries

```python
# Oftentimes you need libraries that are created explicitly to do something you need done.
# This way you don't need to always code everything from scratch.
# An example would be the statistics library
# We can import it like this:

import statistics as stat

# The 'as stat' in this line ensure that you do not continuously have to type 'statistics' but that Python will understand that
# 'stat' means the exact same
# next we can use it like this

students_lengths = [1.63, 1.68, 1.70, 1.81, 1.90]

# median() is a function in the library statistics. By typing stat.median(), Python knows it needs to search for the function
# in the statistics library
print(stat.median(students_lengths))

# In this case the library was already installed for you. Most of the time you need to install it first
```

#### Example time :)

Say we want to create a code that selects all even numbers from a list. It'd look something like this :)

```python
all_numbers = [0, 4, 16, 7, 5, 42, 19, 3, 24, 4, 17, 48, 91]
print('All numbers:', all_numbers) # note that the print is cool enough that it allows you to print several things at once

even_numbers = [] # This is where we will store all even numbers
for item in all_numbers:
    #The '==' is not a typo, it is necessary to see whether two things are exactly the same in Python
    if item % 2 == 0:
        # .append() adds an element to the list whose name is before the "."
        even_numbers.append(item)

print(even_numbers)

# Do you understand what happens in every line of the code? If not, try to add a few print()-statements. This way you'll see
# exactly what's going on and you'll often find what's wrong (also very useful when you're writing your own code).
```

```python
# Sometimes you want to be able to use a specific snippet of code several times, that's when we define a function.
# The code below shows how to define a function of the example above

# To create a function, type "def", then the name of your function, and then two brackets () and :
# If your function needs any input, you can store that information between the brackets

def select_even_numbers(list_of_numbers):
    even_numbers = []
    for item in list_of_numbers:
        if item % 2 == 0:
            even_numbers.append(item)
    return even_numbers # this is the output of the function

# select_even_numbers(all_numbers) is how we 'call' the function. You can do this anywhere in the file
# except for the cells above the cell where you define your function.
# By assigning the function to a variable, the variable even_numbers will contain the output of the function (what is returned)
even_numbers = select_even_numbers(all_numbers)

print('Even numbers:', even_numbers)
```
